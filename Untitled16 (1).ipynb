```python
import cv2
import matplotlib.image as img
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
original_image = img.imread("Resim1.png")
```

```python
flipVertical = cv2.flip(original_image, 0)
flipHorizontal = cv2.flip(original_image, 1)
```

```python
img_rotate_90_clockwise = cv2.rotate(original_image, cv2.ROTATE_90_CLOCKWISE)
img_rotate_90_counterclockwise = cv2.rotate(original_image, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
# Grafikte net ayırt edilebilmesi için yüzde 10 scale ettim.
# Yüzde 50 belli olmuyordu.
resized = cv2.resize(original_image,(0,0),fx=0.1,fy=0.1) 
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3, 2, 1)
plt.imshow(original_image)
plt.axis('off')
plt.title("Original Image")
plt.subplot(3, 2, 2,)
plt.imshow(resized)
plt.axis('off')
plt.title("Resized Image")
plt.subplot(3, 2, 3)
plt.imshow(flipVertical)
plt.axis('off')
plt.title("Flipped Vertically")
plt.subplot(3, 2, 4)
plt.axis('off')
plt.imshow(flipHorizontal)
plt.title("Flipped Horizontally")
plt.subplot(3, 2, 5)
plt.axis('off')
plt.imshow(img_rotate_90_clockwise)
plt.title("Rotated Right")
plt.subplot(3, 2, 6)
plt.axis('off')
plt.imshow(img_rotate_90_counterclockwise)
plt.title("Rotated Left")
print("")
```
